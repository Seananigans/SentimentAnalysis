```python
test = "This is a sentence that I am using to test something. I hope that it works, because otherwise I will be sorely upset."
```

```python
def test_func(test, gram=2):
    if type(test)==str:
        test = test.split()
    list_of_lists = [test[g:] for g in range(gram)]
    return [w for w in zip(*list_of_lists)]
```

```python
test_func(test,3)
```

```python
test_func(text1,3)
```
